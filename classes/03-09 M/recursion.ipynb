````python
def hello_0():
    print('hello')
    hello_0()

hello_0()
````

### What happens here?

Each time `hello_0()` is called, Python creates a **new stack frame**.

Since the function calls itself and never stops, frames keep being added to the stack.

This continues until Python reaches its recursion limit (about **1000 calls by default**) or runs out of memory.

Python then stops the program with:

`RecursionError: maximum recursion depth exceeded`


## Controlling Recursion Depth

We can stop recursion by adding a **parameter** and a **stopping condition**.

````python
def hello_1(n):
    if n == 5: return
    print(n, 'hello')
    hello_1(n+1)

hello_1(0)
````

By adding a parameter (`n`) and a stopping condition (`if n == 3: return`), we control how deep the recursion goes.

STACK

hello_1(3)    
    prints 'hello'
    calls hello_1(3 - 1)

    hello_1(2)       
        prints 'hello'
        calls hello_1(2 - 1)

        hello_1(1)
            prints 'hello'
            calls hello_1(1 - 1)

            hello_1(0)
                n == 0
                returns 
            ↑ returns to hello_1(1)

        hello_1(1)            
        ↑ returns to hello_1(2)

    hello_1(2)        
    ↑ returns to hello_1(3)

hello_1(3)    
↑ returns to main program



## Example with a Global Variable

This version accumulates a total using a global variable.

````python
total = 0
def sum_0(n):
    global total
    if n == 0: return total
    total  = total + n
    print('inside',total)
    
    sum_0(n - 1)

print('outside', sum_0(3))
````

### Stack Diagram
```
STACK

sum_0(3)
    global total updated to 3
    prints 3
    calls sum_0(2)
    waits for sum_0(2) to finish

    sum_0(2)
        global total updated to 5
        prints 5
        calls sum_0(1)
        waits for sum_0(1) to finish

        sum_0(1)
            global total updated to 6
            prints 6
            calls sum_0(0)
            waits for sum_0(0) to finish

            sum_0(0)
                n == 0
                returns 6
            ↑ returns to sum_0(1)

        sum_0(1)
            does not return the value
            function ends → returns None
        ↑ returns to sum_0(2)

    sum_0(2)
        function ends → returns None
    ↑ returns to sum_0(3)

sum_0(3)
    function ends → returns None
↑ returns to main program

main program prints:
outside None
```


## Correct Recursive Pattern

Instead of using a global variable, we return values up the stack.

### Recurrence

`sum_1(n) = n + sum_1(n-1)`

### Base Case

`sum_1(0) = 0`


````python
def sum_1(n):

    if n == 0 : return 0

    return n + sum_1(n - 1)

print(sum_1(3))
````

### Stack Diagram
```
STACK

sum_1(3)
    returns 3 + sum_1(2)
    waits for sum_1(2)

    sum_1(2)
        returns 2 + sum_1(1)
        waits for sum_1(1)
        
        sum_1(1)
            returns 1 + sum_1(0)
            waits for sum_1(0)
            
            sum_1(0)
                returns 0
            ↑ returns 0 to sum_1(1)

        sum_1(1)
            returns 1 + 0 = 1
        ↑ returns 1 to sum_1(2)

    sum_1(2)
        returns 2 + 1 = 3
    ↑ returns 3 to sum_1(3)

sum_1(3)
    returns 3 + 3 = 6
↑ returns 6 to main program

main program prints:
6
```

### Two Important Notes

1. We do **not** use a global variable.
   The function communicates only through its parameters and return values.
2. Each recursive call creates its own stack frame.
   Each frame has its own local variables, so each call has its own value of `n`.
